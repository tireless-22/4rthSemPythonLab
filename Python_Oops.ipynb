```python
class Rectangle:
    name="rectangle"
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth

    def area(self):
        area=self.length*self.breadth
        return area

    def perimeter(self):
        perimeter=2*(self.length+self.breadth)
        return perimeter

r1=Rectangle(20,10)


print(r1.length)
print(r1.breadth)
print(r1.area())
print(r1.perimeter())
```

<h1>Method Types</h1>
<p>
The three types of Methods in OOP are:
</p>
<pre>
1. Instance methods
2. Class methods
3. Static methods
</pre>
<p>
<b>
Instance methods:
</b></p>
<p>
These are the methods used to get or set object attributes. These methods are also known as accessor methods (gets the details), mutator methods (sets the details). In our example, the “show” method is the getter/accessor method and the “update” method is the setter/mutator method.
</p>

<p>
<b>
Class methods:
</b></p>
<p>
Class methods are bound to the class rather than an instance. They can access and modify class-level attributes but not instance-specific data.
</p>

<p>
<b>
Static Methods:
</b></p>
<p>
Static methods are not bound to the instance or the class and do not have access to instance or class attributes. They are mainly used for utility functions that don't depend on the state of the instance or class.
</p>


```python
class MyClass:
    class_variable = "I am a class variable"

    def __init__(self, instance_variable):
        self.instance_variable = instance_variable

    def instance_method(self):
        print("instance variable",self.instance_variable)

    @classmethod
    def class_method(cls):
        print("Class method called. Class variable",cls.class_variable)

    @staticmethod
    def static_method():
        print("Static method called.")

# Creating an instance of MyClass
obj = MyClass("I am an instance variable")

# Calling instance method
obj.instance_method()

# Calling class method
MyClass.class_method()

# Calling static method
MyClass.static_method()
```

<h1>Inner classes</h1>
<p>
A class defined in another class is known as an inner class or nested class. If an object is created using child class means inner class then the object can also be used by parent class or root class. A parent class can have one or more inner classes but generally inner classes are avoided.
</p>

```python
class OuterClass:
    def __init__(self, outer_name):
        self.outer_name = outer_name
        self.inner_instance = self.InnerClass()  # Creating an instance of the inner class

    def display_outer_name(self):
        print("Outer name:",self.outer_name)

    class InnerClass:
        def display_inner_message(self):
            print("Hello from the InnerClass!")

# Creating an instance of the OuterClass
outer_obj = OuterClass("OuterInstance")

# Accessing and displaying outer instance variable
outer_obj.display_outer_name()

# Accessing inner class method through outer class instance
outer_obj.inner_instance.display_inner_message()
```

<h1>Inheritance</h1>
<p>The capability of a class to derive properties and characteristics from another class is called Inheritance. Inheritance is one of the most important feature of Object Oriented Programming.</p>


*   <b>Sub Class:</b> The class that inherits properties from another class is called Sub class or Derived Class.
*   <b>Super Class:</b> The class whose properties are inherited by sub class is called Base Class or Super class.
*   <b>Reusability:</b> Inheritance supports the concept of “reusability”, i.e. when we want to create a new class and there is already a class that includes some of the code that we want, we can derive our new class from the existing class. By doing this, we are reusing the fields and methods of the existing class.

image.png

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def make_sound(self):
        print(self.name," says ",self.sound)

class Dog(Animal):
    def wag_tail(self):
        print(self.name," is wagging its tail")

class Cat(Animal):
    def scratch_furniture(self):
        print(self.name, "is scratching the furniture")

class Cow(Animal):
    def chew_cud(self):
        print(self.name," is chewing its cud")

# Instances of the classes
dog = Dog("Buddy", "Woof")
cat = Cat("Whiskers", "Meow")
cow = Cow("Bessie", "Moo")

# Using methods of the classes
dog.make_sound()
dog.wag_tail()

cat.make_sound()
cat.scratch_furniture()

cow.make_sound()
cow.chew_cud()
```

<h1>super()</h1>
<p>
In Python, super() is a built-in function that is commonly used in object-oriented programming (OOP) to refer to the parent class. It is used to call a method from the parent class and is often used in the context of inheritance.
</p>

```python
class Parent:
    def __init__(self, name):
        self.name = name

    def show_info(self):
        print("Parent class: ",self.name)

class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age

    def show_info(self):
        super().show_info()
        print("Child class:", self.age)

# Creating an instance of the Child class
child_obj = Child("Alice", 10)

# Calling the show_info method of the Child class
child_obj.show_info()
```

```python
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def make_sound(self):
        print(self.name," says ",self.sound)

class Dog(Animal):
    def __init__(self, name):
        # Calling the constructor of the parent class
        super().__init__(name, "Woof")

    def wag_tail(self):
        print(self.name," is wagging its tail")

class Cat(Animal):
    def __init__(self, name):
        super().__init__(name, "Meow")

    def scratch_furniture(self):
        print(self.name," is scratching the furniture")

class Cow(Animal):
    def __init__(self, name):
        super().__init__(name, "Moo")

    def chew_cud(self):
        print(self.name," is chewing its cud")

# Creating instances of the classes
dog = Dog("Buddy")
cat = Cat("Whiskers")
cow = Cow("Bessie")

# Using methods of the classes
dog.make_sound()
dog.wag_tail()

cat.make_sound()
cat.scratch_furniture()

cow.make_sound()
cow.chew_cud()
```
